```python
class Node:
    
    def __init__(self,data=None):
        self.data=data
        self.next=None
        
    def setData(self,data):
        self.data=data
        
    def getData(self):
        return self.data
    
    def setNext(self,node):
        self.next=node
        
    def getNext(self):
        return self.next
    
    def hasNext(self):
        return self.next != None
    
class LinkedList:
    
    def __init__(self,node=None):
        self.head = node
        self.h=defaultdict(int)
        
    def lengthList(self):
        temp = self.head
        c=0
        while temp:
            c+=1
            temp=temp.next
        return c
        
    def addNode(self,data,pos):
        if self.head==None and pos==1:
            n=Node(data)
            self.head=n
        else:
            if pos==1:
                n=Node(data)
                n.next=self.head
                self.head=n
            elif pos==self.lengthList()+1:
                cur=self.head
                while cur.next:
                    cur=cur.next
                n=Node(data)
                cur.next=n  
            else:
                p=None
                cur=self.head
                while pos and cur.next:
                    p=cur
                    cur=cur.next
                    pos-=1
                n=Node(data)
                n.next=p.next
                p.next=n
        self.h[n]=n
        
    def printAllElements(self):
        t=self.head
        while t:
            print(t.data)
            t=t.next
            
    def deleteNode(self,pos):
        if self.head==None and pos==1:
            print("Empty Linked list")
        else:
            if pos==1:
                t=self.head
                self.head=t.next
                del self.h[t]
            elif pos==self.lengthList():
                t=self.head
                p=None
                while t.next:
                    p=t
                    t=t.next
                del self.h[t]
                p.next=None
            else:
                p=None
                t=self.head
                while pos > 1 and t:
                    p=t
                    t=t.next.next
                    pos-=1
                del self.h[p.next]
                p.next=t 
```

```python
from collections import defaultdict
if __name__=="__main__":
    #Creating Linked list using text file.
    # Text file contains 1-10
    l1=LinkedList()
    c=1
    with open('/home/uma/input.txt') as file:
        for x in file.readlines():
            l1.addNode(int(x.rstrip('\n')),c)
            c+=1
           
    #print all element
    
    print("Printing all elements : {}".format(l1.printAllElements()))
    
    #adding the element at the end
    l1.addNode(11,l1.lengthList()+1)
    print("Printing all elements after adding at the end : {}".format(l1.printAllElements()))
    ###adding at the beginning
    l1.addNode(12,1)
    print("Printing all elements after adding at the beginning : {}".format(l1.printAllElements()))
    ####adding randomly 
    l1.addNode(34,5)
    print("Printing all elements after adding randomly : {}".format(l1.printAllElements()))
    #####delete at beginning
    l1.deleteNode(1)
    print("Printing all elements after deleting at the beginning : {}".format(l1.printAllElements()))
    ####delete at end
    l1.deleteNode(l1.lengthList()-1)
    print("Printing all elements after deleting at the end : {}".format(l1.printAllElements()))
    ### delete randomly
    l1.deleteNode(3)
    print("Printing all elements after deleting randomly: {}".format(l1.printAllElements()))
    l1.printAllElements()
    print("Printing elements in hash table")
    for i in l1.h:
        print(i.data)
```
